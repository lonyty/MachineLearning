```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics
plt.rcParams['font.size'] = 14
plt.style.use('seaborn-whitegrid')
```

```python
df = pd.read_csv('cleaned_data/Cleaned Data.csv',index_col=[0])
```

```python
df
```

```python
df['Signal'].value_counts()
```

```python
SignalFac = []

for string in df['Signal']:
    if string == 'Sell':
        SignalFac.append(0)
    elif string == 'Hold':
        SignalFac.append(1)    
    else:
        SignalFac.append(2)
```

```python
df['SignalFac'] = SignalFac
df.drop('Signal', axis=1, inplace=True)
```

```python
df
```

```python
X = df.drop('SignalFac',axis=1)
y = df['SignalFac']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

# Create LDA object and run classifier
lda = LDA(solver='lsqr')
lda = lda.fit(X_train, y_train)
lda

# Relabel variables as discussed
X_test = X_train
y_test = y_train

# Predict labels
y_pred = lda.predict(X_test)

# Sklearn's confusion matrix
print(metrics.confusion_matrix(y_test, y_pred))
```

```python
print('default-rate: {0: .4f}'.format(np.sum(y_train)/len(y_train)))
print('score:        {0: .4f}'.format(lda.score(X_train, y_train)))
print('error-rate:   {0: .4f}'.format(1-lda.score(X_train, y_train)))
```

## Lösen der Class Imbalance mit Upscaling

```python
#Hold werden upgesampled
#Sell werden downgesampled
```

```python
print('Sell hat', X[y==0].shape[0], 'Werte')
print('Hold hat', X[y==1].shape[0], 'Werte')
print('Buy hat', X[y==2].shape[0], 'Werte')
```

```python
from sklearn.utils import resample

X_upsampled, y_upsampled = resample(X[y==1], y[y==1],
                                    replace=True,
                                    n_samples=X[y==2].shape[0],
                                    random_state=1)
```

```python
print(y_upsampled)
```

```python
print('Sell hat', X[y==0].shape[0], 'Werte')
print('Hold hat', X[y==1].shape[0], 'Werte')
print('Buy hat', X[y==2].shape[0], 'Werte')
```

```python
# Downsampling
X_dnsampled, y_dnsampled = resample(X[y==0], y[y==0],
                                    replace=False,
                                    n_samples=X[y==2].shape[0],
                                    random_state=1)
```
