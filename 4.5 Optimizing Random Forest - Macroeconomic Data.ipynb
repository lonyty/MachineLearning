```python
df2014 = pd.read_csv('raw_data/2014_Financial_Data.csv')
df2015 = pd.read_csv('raw_data/2015_Financial_Data.csv')
df2016 = pd.read_csv('raw_data/2016_Financial_Data.csv')
df2017 = pd.read_csv('raw_data/2017_Financial_Data.csv')
df2018 = pd.read_csv('raw_data/2018_Financial_Data.csv')
```

```python
df2014 = df2014.rename({'2015 PRICE VAR [%]':'PRICE_VAR'},axis =1)
df2015 = df2015.rename({'2016 PRICE VAR [%]':'PRICE_VAR'},axis =1)
df2016 = df2016.rename({'2017 PRICE VAR [%]':'PRICE_VAR'},axis =1)
df2017 = df2017.rename({'2018 PRICE VAR [%]':'PRICE_VAR'},axis =1)
df2018 = df2018.rename({'2019 PRICE VAR [%]':'PRICE_VAR'},axis =1)
```

```python
# Adding the Benchmark for each year
df2014['S&P 500'] = -0.0073
df2015['S&P 500'] = 0.0954
df2016['S&P 500'] = 0.1942
df2017['S&P 500'] = -0.0624
df2018['S&P 500'] = 0.2888
```

```python
for colname in col:

    df2018[colname] = df_macrodata[colname][0]
    df2017[colname] = df_macrodata[colname][1]
    df2016[colname] = df_macrodata[colname][2]
    df2015[colname] = df_macrodata[colname][3]
    df2014[colname] = df_macrodata[colname][4]
```

```python
df2018
```

```python
print(df2014.shape)
print(df2015.shape)
print(df2016.shape)
print(df2017.shape)
print(df2018.shape)
```

```python
df = pd.concat([df2014,df2015,df2016,df2017,df2018])
df['PRICE_VAR']=df['PRICE_VAR']/100
```

```python
#create index
df.index = np.arange(1,22078) 
```

```python
df = df.drop(['Unnamed: 0'], axis=1)
```

```python
#Threshhold
Threshhold = 22077*(0.5)
df = df.dropna(thresh=Threshhold, axis=1)
df = df.dropna(thresh=220*0.5-2)
```

```python
#Creating dummy variables
sector = df['Sector'] 
sector_dummies = pd.get_dummies(sector)
df=df.drop('Sector', axis=1)
```

```python
#Estimation missing data points
df = df.fillna(value=df.median())
```

```python
#Removing outliers
df = df[np.abs(df-df.mean()) <= (2.33*df.std())].dropna()
df.shape
```

```python
df = df.join(sector_dummies)
```

```python
#Adding classification buy/hold/sell
```

```python
dif = df['PRICE_VAR'] - df['S&P 500']

signal = []

for number in dif:
    if number<-0.025:
        signal.append('Sell')
    elif number<0.025:
        signal.append('Hold')       
    else:
        signal.append('Buy')
```

```python
df['Signal']= signal
```

```python
df = df.drop(['S&P 500','Class','PRICE_VAR'], axis=1) 
```

```python
df.index = np.arange(1,11593) 
```

## Decision Trees

```python
#Assigning values
X = df.drop('Signal', axis=1)
y = df['Signal']
```

```python
# Create Train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0, stratify=y)
```

```python
#scale our data to avoid issues (X_scaled = X_std * (max - min) + min)
from sklearn.preprocessing import MinMaxScaler
```

```python
#creating an instance
scaler = MinMaxScaler()

#adjusting scaler only based on the training set (without test values to prevent data leakage)
scaler.fit(X_train)
```

```python
#perform the transformation
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Create classifier object and fit it to data
forest = RandomForestClassifier(criterion='gini', random_state=0, n_jobs=-1, max_depth=10)
forest.fit(X_train, y_train)
```

```python
y_pred = forest.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn import metrics
print('Confusion matrix: \n', 
      metrics.confusion_matrix(y_test, y_pred))
```

```python
#Cross Validation

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = \
    train_test_split(X, y, 
                     test_size=0.3, 
                     random_state=0, 
                     stratify=y)

# Import necessary functions
from sklearn.model_selection import StratifiedKFold, cross_val_score

# Create k-Fold CV and LogReg object
kFold = StratifiedKFold(n_splits=5)
forestcv = RandomForestClassifier(criterion='gini', random_state=0, n_jobs=-1)

# Run CV and print results
scores = cross_val_score(forestcv, X_train, y_train, cv=kFold)
print(scores)
print('CV accuracy on train set: {0: .3f} +/- {1: .3f}'.format(np.mean(scores), np.std(scores)))
```
