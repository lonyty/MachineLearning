# Data cleaning

We see that our data set brings with it various problems that we have to look at more closely. We noticed the following:
1. The dataset are missing some data (Depending on the feature, it is more or less that we are missing)
2. Replace string into dummy indicators
3. We still have to add an estimated value to missing data points
4. Remove outliers
5. We need to create the classification indicator Buy/Hold/Sell

## 1. Checking missing data:

* 1 step: Analyzing columns
* 2 step: Analyzing rows

### 1 Step: Analyzing columns

```python
sns.heatmap(df.isnull(), yticklabels=False,cbar=False,cmap='viridis',xticklabels='')

plt.ylabel('Rows')
plt.title('Columns')

#plt.savefig('DataFrame.png',dpi=200)
```

```python
# Our threshhold indicates how many values a feature need to be considered
# We set our threshhold at 50% of needed values, that are not NaN

Threshhold = 22077*(0.5)
print('Our treshhold is', round(Threshhold, 0), 'data points')
```

```python
df = df.dropna(thresh=Threshhold, axis=1)
```

```python
sns.heatmap(df.isnull(), yticklabels=False, cbar=False,cmap='viridis')
```

```python
df.shape
```

### 2 Step: Analyzing rows

```python
# Additionally we can see in our heatmap that for some rows most of its features are missing
# We also set the threshhold at 50% of features that a row needs to proceed

df = df.dropna(thresh=220*0.5-2)
df.shape
```

```python
sns.heatmap(df.isnull(), yticklabels=False, cbar=False,cmap='viridis')
```

## 3. Add estimated values to missing data points

```python
# We will fill missing data points with the median
# This is because our dataframe stills includes some outliers that we will clean afterwards
```

```python
df = df.fillna(value=df.median())
df.info()
```

```python
sns.heatmap(df.isnull(), yticklabels=False, cbar=False,cmap='viridis')
```

## 4. Removing Outliers

* Outliers need to be removed from our dataset because it would have a strong influence in our predictions 
* On kaggle, the autor mentions that the 1% - 99% percentile could incluede extremas 
* We came up with the following plan
* We remove rows that inclue data points that are above or below the 2.33 standard deviation 
* This approach is a more secure way to remove all corrupt data

```python
from IPython.display import Image
Image(filename='images/Normalverteilung.png')
```

```python
df
```

```python
#https://stackoverflow.com/questions/23199796/detect-and-exclude-outliers-in-pandas-data-frame

df = df[np.abs(df-df.mean()) <= (2.33*df.std())].dropna()
```

## Adding classification buy/hold/sell

* Buy: Outperformance 2.5% compared to the S&P500
* Hold: Performance between +/- 2.5% compared to the S&P 500
* Sell: Underperformance of 2.5% compared to the S&P 500

```python
dif = df['PRICE_VAR'] - df['S&P 500']
dif
```

```python
signal = []

for number in dif:
    if number<-0.025:
        signal.append('Sell')
        print('Sell')
    elif number<0.025:
        signal.append('Hold')
        print('Hold')
    else:
        signal.append('Buy')
        print('Buy')
```

```python
df['Signal']= signal
```

```python
sns.countplot(df['Signal'],palette='muted')
plt.title('Class Distribution')
```
